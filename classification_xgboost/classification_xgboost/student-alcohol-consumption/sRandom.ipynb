```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,accuracy_score,recall_score,confusion_matrix
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataSet1 = pd.read_csv('student-por.csv')
dataSet1.head()
```

```python
dataSet2 = pd.read_csv('student-mat.csv')
dataSet2.head()
```

```python
dataSet=pd.concat([dataSet1,dataSet2], ignore_index=True)
dataSet.head()
```

```python
target=dataSet['Walc']
dataSet.drop('Walc',axis=1,inplace=True)
dataSet.drop('Dalc',axis=1,inplace=True)
target.head()
```

```python
dataSet=pd.get_dummies(dataSet)
dataSet.head()
```

```python
x=dataSet
y=target
```

```python
#train and test data set spilitting
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.30, random_state=42)
```

```python
from xgboost import XGBClassifier
classifier = XGBClassifier()
classifier.fit(X_train, y_train)
```

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
```

```python
# Applying k-Fold Cross Validation
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv = 10)
print("accuracy{:.3f}".format(accuracies.mean()))
#accuracies.std()
```

```python
# Making the Confusion Matrix
a=confusion_matrix(y_test, y_pred)
a
```
